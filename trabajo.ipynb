```python
# ============================================
# LABORATORIO DE ANÁLISIS NUMÉRICO NO LINEAL
# TALLER – INTEGRACIÓN NUMÉRICA
# ============================================

import numpy as np
import math
from scipy import integrate



def f(x):
    return math.exp(-x) * math.cos(5*x)

def g(x):
    return math.sqrt(x * math.log(x + 1))

a_f, b_f = 0.0, 2.0
a_g, b_g = 0.0, 1.0



def punto_medio(f, a, b, n):
    h = (b - a) / n
    s = 0.0
    for i in range(n):
        s += f(a + (i + 0.5) * h)
    return h * s

def trapecio(f, a, b, n):
    h = (b - a) / n
    s = 0.5 * (f(a) + f(b))
    for i in range(1, n):
        s += f(a + i*h)
    return h * s

def simpson(f, a, b, n):
    h = (b - a) / n
    s = f(a) + f(b)
    for i in range(1, n):
        s += (4 if i % 2 else 2) * f(a + i*h)
    return h * s / 3



print("\nPREGUNTA 1.a")
print("f(x) = e^{-x}cos(5x) pertenece a C^∞([0,2]) porque")
print("es producto de funciones suaves (exponencial y coseno).")

print("\nPREGUNTA 1.b")
print("La integral existe y es única porque f es continua en [0,2].")


If_ref, _ = integrate.quad(f, a_f, b_f, epsabs=1e-12)

print("\nPREGUNTA 2")
print("Valor de referencia If =", If_ref)


print("\nPREGUNTAS 3 y 4")

for n in [4, 8]:
    print(f"\nSubintervalos n = {n}")
    pm = punto_medio(f, a_f, b_f, n)
    tr = trapecio(f, a_f, b_f, n)
    si = simpson(f, a_f, b_f, n)

    print(" Punto medio:", pm, " Error:", abs(pm - If_ref))
    print(" Trapecio   :", tr, " Error:", abs(tr - If_ref))
    print(" Simpson    :", si, " Error:", abs(si - If_ref))


E4_tr = abs(trapecio(f, a_f, b_f, 4) - If_ref)
E8_tr = abs(trapecio(f, a_f, b_f, 8) - If_ref)
orden_tr = math.log(E4_tr / E8_tr, 2)

E4_si = abs(simpson(f, a_f, b_f, 4) - If_ref)
E8_si = abs(simpson(f, a_f, b_f, 8) - If_ref)
orden_si = math.log(E4_si / E8_si, 2)

print("\nPREGUNTA 5")
print("Orden empírico trapecio:", orden_tr)
print("Orden empírico Simpson :", orden_si)
print("Simpson es mejor por su orden teórico 4.")


print("\nPREGUNTA 6")
print("Simpson n=16 :", simpson(f, a_f, b_f, 16))
print("Trapecio n=16:", trapecio(f, a_f, b_f, 16))

print("\nPREGUNTA 7")
print("Simpson es más eficiente.")
print("La oscilación de cos(5x) afecta más al trapecio.")

R = np.zeros((3,3))
R[0,0] = trapecio(f, a_f, b_f, 1)
R[1,0] = trapecio(f, a_f, b_f, 2)
R[2,0] = trapecio(f, a_f, b_f, 4)

R[1,1] = R[1,0] + (R[1,0] - R[0,0]) / 3
R[2,1] = R[2,0] + (R[2,0] - R[1,0]) / 3
R[2,2] = R[2,1] + (R[2,1] - R[1,1]) / 15

print("\nPREGUNTA 8")
print("Tabla de Romberg:")
print(R)
print("R2,2 =", R[2,2])

print("\nPREGUNTA 9")
print("La extrapolación de Richardson requiere suavidad.")
print("Romberg es más costoso que Simpson.")



If_adapt, err, info = integrate.quad(f, a_f, b_f, epsabs=1e-6, full_output=1)

print("\nPREGUNTA 10")
print("Valor por método adaptativo:", If_adapt)

print("\nPREGUNTA 11")
print("Número de evaluaciones:", info['neval'])

print("\nPREGUNTA 12")
print("El método adaptativo refina automáticamente")
print("las regiones donde la función es más problemática.")



Ig_ref, _ = integrate.quad(g, a_g, b_g)

print("\nPREGUNTA 13.a")
print("g(x) no pertenece a C^2([0,1]) por el comportamiento en x=0.")

print("\nPREGUNTA 13.b")
print("Trapecio :", trapecio(g, a_g, b_g, 16))
print("Simpson  :", simpson(g, a_g, b_g, 16))
print("Adaptativo:", Ig_ref)

print("\nPREGUNTA 13.c")
print("Simpson pierde eficiencia por falta de suavidad.")
print("El método adaptativo mantiene precisión.")


print("\nPREGUNTA 14")
print("Funciones suaves: Simpson")
print("Funciones oscilatorias: Romberg o adaptativos")
print("Funciones con singularidades: métodos adaptativos")
print("No existe un método universalmente óptimo.")
```
